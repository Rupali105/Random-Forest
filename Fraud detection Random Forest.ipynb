# Random Forest : Fraud detection

#Use Random Forest to prepare a model on fraud data

#treating those who have taxable_income <= 30000 as "Risky" and others are "Good"

```python
import pandas as pd
import numpy as np
import category_encoders as ce
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error, classification_report
import matplotlib.pyplot as plt
from sklearn import tree
import seaborn as sns

from sklearn.preprocessing import StandardScaler
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['figure.dpi'] = 1000
sns.set_theme(style='darkgrid', palette='rainbow')
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Fraud_check.csv')
df
```

```python

encoder=ce.OrdinalEncoder(cols=['Undergrad', 'Marital.Status', 'Urban'])
data1=encoder.fit_transform(df)
data1
```

```python
tax_val = []
for value in df["Taxable.Income"]:
    if value<=30000:
        tax_val.append("Risky")
    else:
        tax_val.append("Good")
        
data1["tax_val"]= tax_val
```

```python
data1
```

```python
x = data1.drop(['tax_val','Taxable.Income'], axis =1)
y = data1['tax_val']
x
```

```python
sc = StandardScaler()

x = sc.fit_transform(x)
x
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=40)
```

# Building a Randon forest classifier

```python
rfc = RandomForestClassifier()
rfc.fit(xtrain,ytrain)
yp = rfc.predict(xtest)
```

```python
yp
```

```python
print(classification_report(ytest,yp))
```
